```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Importing Dataset

```python
df = pd.read_csv('Algerian_forest_fires_dataset.csv',header = 1)
df.head()
```

# Data Cleaning

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.drop(df.iloc[122:124].index , inplace = True)
df.reset_index(inplace = True)
```

```python
df.head(4)
```

```python
df.drop('index',axis = 1 , inplace = True)
```

```python
df.head(5)
```

```python
df.loc[: 122 , 'region'] ='bejaia'
df.loc[122 : , 'region'] = 'Sidi-Bel Abbes'
```

```python
df.head(4)
```

```python
df.tail()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
def info():
    
    data = pd.DataFrame(index=df.columns)
    data['Data_Type'] = df.dtypes
    data['Null_count'] = df.isnull().sum()
    data['unique_count'] = df.nunique()
    data['Duplicate_count'] = df.duplicated().sum()
    return data
```

```python
info()
```

```python
df['Classes'].unique()
```

```python
df['Classes'] = df['Classes'].str.strip()
```

```python
df['Classes'].unique()
```

```python
df[df['Classes'].isnull()]
```

```python
df['FWI'] = df['FWI'].apply(lambda x : x.replace('fire' , ' 0.0 '))
```

```python
df[df['Classes'].isnull()]
```

```python
df['Classes'].unique()
```

```python
modeValueRating = df['Classes'].mode()
```

```python
modeValueRating
```

```python
df['FWI'] = df.FWI.replace(r' ^\s*$', np.nan , regex  = True)
```

```python
df['FWI'].mode()
```

```python
df['Classes'].fillna(value=modeValueRating[0],inplace = True)
df['FWI'].fillna(df['FWI'].mode(),inplace  = True)
df['DC'] = df.DC.apply(lambda x : x.replace(' ',''))
```

```python
df
```

```python
info()
```

```python
## Changing the Data type of the following
df['Temperature'] = df['Temperature'].astype(int)
df['RH'] = df['RH'].astype(int)
df['Rain'] = df['Rain'].astype(float)
df['FFMC'] = df['FFMC'].astype(float)
df['DMC'] = df['DMC'].astype(float)
df['BUI'] = df['BUI'].astype(float)
df['ISI'] = df['ISI'].astype(float)
df['Ws'] = df['Ws'].astype(float)
df['DC'] = df['DC'].astype(float)
df['FWI'] = df['FWI'].astype(float)
```

```python
info()
```

# Copy the Clean Dataset

```python
df.to_csv("Algerian_forest_fire_cleaned_data.csv" , index = False)
```

# Univariate Analysis

The term univariate analysis refer to the analysis of one variable prefix "uni" means "one". The purpose of the univariate analysis is to understand the distribution of value for a single variable

```python
## Numerical Feature
```

```python
plt.figure(figsize = (15,20))
plt.suptitle('Univariate Analysis of the Numerical Feature' , fontsize = 20 , fontweight='bold',alpha = 0.8 , y = 1)


for i in range(0 , len(numerical_feature)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x = df[numerical_feature[i]],shade = True , color = 'g' , hue = 'region' , data = df)
    plt.xlabel(numerical_feature[i])
    plt.tight_layout()
```

Observation : 

1. In the temperature , RH ,and WS is normally distributed

2. Rain is normally distributed with some outliear

3. FFMC is left skewed 

4. DMC , DC , ISI , BUI and FWI is log normal distributed with the data.

# Label Encoding

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Classes'] = label_encoder.fit_transform(df['Classes'])
df.head()
```

```python
df['Classes'].unique()
```

```python
X = df.iloc[:,3:13]
y = df['Classes']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
```

```python
X_train = Scaler.fit_transform(x_train)
X_test = Scaler.transform(x_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
sns.histplot(X_train)
```

```python
sns.histplot(x_train)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
logistic_regression
```

```python
logistic_regression.fit(X_train , y_train)
```

```python
print("Cofficent of the feature  : " , logistic_regression.coef_)
```

```python
print("Intercept of the feature : " , logistic_regression.intercept_)
```

```python
logistic_regression_predict = logistic_regression.predict(X_test)
logistic_regression_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Training Score :",logistic_regression.score(X_train,y_train))
print("Test Score : " , logistic_regression.score(X_test,y_test))
```

```python
accuracy = accuracy_score(y_test,logistic_regression_predict)
accuracy
```

```python
## Classification Report
```

```python
from sklearn.metrics import classification_report
```

```python
score = accuracy_score(y_test,logistic_regression_predict)
Classification_Report = classification_report(y_test,logistic_regression_predict)

print("Logestic Regression")

print(Classification_Report)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
Confusion_Matrix = ConfusionMatrixDisplay.from_estimator(logistic_regression,X_test,y_test)
Confusion_Matrix
```
